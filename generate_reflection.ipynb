```python
reflection_path = '/shared/data3/yichi9/hotpot_dev_reflections.jsonl'
reflection_model = "/shared/data3/yichi9/Mistral-7B-Instruct-v0.2-sayself-question-full"

cuda_devices = "5"
config_path = "./hotpotqa_full.yaml"
dataset_split = "dev"
dataset_query_key = "question"
```

```python
from flashrag.config import Config
import os

config = Config(config_file_path=config_path)
os.environ['CUDA_VISIBLE_DEVICES'] = cuda_devices

from flashrag.utils import get_dataset
dataset = get_dataset(config)
```

```python
from sayrag import generate_reflections
from vllm import LLM
from vllm import SamplingParams
from vllm.lora.request import LoRARequest

sampling_params = SamplingParams(temperature=0.0, max_tokens=1024)


llm = LLM(model=reflection_model, 
            max_model_len=2048,
            gpu_memory_utilization=0.85 )
```

```python
reflections = generate_reflections(llm, sampling_params, 
                    dataset, split=dataset_split, key=dataset_query_key,
                    output=reflection_path, max_reroll = 1)
```

```python
from sayrag import read_jsonl, write_jsonl

reflections = read_jsonl(reflection_path)
```

```python
from flashrag.utils import get_retriever

retriever = get_retriever(config)
retriever.use_cache = False

reflection_texts = [ref['extra_query'] for ref in reflections]
retrieval_results = retriever.batch_search(reflection_texts)
for ref, r in zip(reflections, retrieval_results):
    ref['retrieval_result'] = r
```

```python
write_jsonl(reflections, reflection_path)
```
